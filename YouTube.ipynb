```python
import requests
from bs4 import BeautifulSoup


search=input("請輸入搜尋名稱: ")
url = "https://www.youtube.com/results?search_query=%s"%search

request = requests.get(url)
soup = BeautifulSoup(request.text, "html.parser")

for mv in soup.select(".yt-lockup-video"):

    data = mv.select("a[rel='spf-prefetch']")
    print("名稱: {}".format(data[0].get("title")))
    print("網址: https://www.youtube.com{}".format(data[0].get("href")))

    data = mv.select(".yt-lockup-meta-info")
    time = data[0].get_text("#").split("#")[0]
    see = data[0].get_text("#").split("#")[1]
    print("發佈時間: {}".format(time))
    print(see)

    data = mv.select("a[rel='spf-prefetch']")
    img = mv.select("img")
    if img[0].get("src") != "/yts/img/pixel-vfl3z5WfW.gif":
        print("照片: {}".format(img[0].get("src")))
    else:
        print("照片: {}".format(img[0].get("data-thumb")))
    print(' ')
```
